Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House():
  def __init__(self,area,rooms,bathrooms,floors,location,garage,garden=False):
    self.area = area
    self.rooms = rooms
    self.bathrooms = bathrooms
    self.floors = floors
    self.garage = garage
    self.garden = garden
    self.location = location

  def calc_price(self):
    #HUF
    price = self.area*1500000

    if self.garage:
      price += 8000000 *self.garage

    if self.garden:
      price += 5000000*self.garden

    return price

  def price_EUR(self):
    return self.calc_price()*0.0028

  def display_info(self):
    print("Information about the house:")
    print("----------------------------")
    print(f"Area: {self.area} m2")
    print(f"Number of rooms: {self.rooms}")
    print(f"Number of bathrooms: {self.bathrooms}")
    print(f"Number of floors: {self.floors}")
    print(f"Number of garages: {self.garage}")
    print(f"Has garden: {self.garden}")
    print(f"Location: {self.location}")
    print(f"Estimated Price: {self.calc_price():,} HUF")
    print(f"Estimated Price: {self.price_EUR():,} EUR")

```

Create a Person class with the following attributes:


*   name
*   age
*   married (T/F)
*   number of children
*   number of cars
*   total amount of money (to buy a house)
*   based on the number of children store the number of family members in an attribute  (- this does not have to be included in the init arguments as it is a calculated value)


(4 points)




```python
class Person():
  def __init__(self,name,age,married,children,cars,money):
    self.name = name
    self.age = age
    self.married = married
    self.children = children
    self.cars = cars
    self.money = money

    if self.married:
      self.family_members = 2 + self.children
    else:
      self.family_members = 1 + self.children

  def print_info(self):
    print("Information about the Person:")
    print("-----------------------------")
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Married: {self.married}")
    print(f"Number of children: {self.children}")
    print(f"Number of cars: {self.cars}")
    print(f"Total amount of money: {self.money} HUF")
    print(f"Number of family members: {self.family_members}")
```

###A) (5 points)
 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

```python
class Animal():
  def __init__(self, name, age, weight):
    self.name = name
    self.age = age
    self.weight = weight

  def swim(self):
    print(f"{self.name} swims in the ocean.")

  def eat(self):
    print(f"{self.name} eats food from the sea.")

  def make_sound(self):
    print(f"{self.name} makes a sound like the sea animals.")

  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Weight: {self.weight}")

class Shark(Animal):
  def __init__(self, name, age, weight, speed, teeth_nr):
    super().__init__(name, age, weight)
    self.speed = speed
    self.teeth_nr = teeth_nr

  def eat(self):
    print(f"{self.name} hunts the fish with {self.teeth_nr} sharp teeth.")

  def swim(self):
    print(f"{self.name} swims at {self.speed} km/h.")

  def make_sound(self):
    print(f"{self.name} is mostly silent.")

  def detect_blood(self):
    print(f"{self.name} detects the blood from far away.")

class Dolphin(Animal):
  def __init__(self, name, age, weight, intelligence):
    super().__init__(name,age,weight)
    self.intelligence = intelligence

  def make_sound(self):
    print(f"{self.name} clicks and whistles to communicate.")

  def swim(self):
    print(f"{self.name} jumps high above the waves during swimming.")

  def tricks(self):
    print(f"{self.name} performs a trick with an intelligence of {self.intelligence}.")

class Octopus(Animal):
  def __init__(self,name,age,weight,arms):
    super().__init__(name,age,weight)
    self.arms = arms

  def swim(self):
    print(f"{self.name} moves by pushing water through his body.")

  def make_sound(self):
    print(f"{self.name} does not make loud sounds.")

  def camouflage(self):
    print(f"{self.name} changes color to hide from enemies.")

  def nrarms(self):
    print(f"{self.name} has {self.arms} arms.")
```

```python
shark1 = Shark("Bruce",120,900,50,300)
dolphin1 = Dolphin("Luna",8,150,10)
octopus1 = Octopus("Ottó",43,20,8)
```

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math

class Shape:
  def area(self):
    pass

class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius

  def area(self):
    ar = math.pi*self.radius**2
    return ar

class Square(Shape):
  def __init__(self,side):
    self.side = side

  def area(self):
    ar = self.side**2
    return ar

class Triangle(Shape):
  def __init__(self,base,height):
    self.base = base
    self.height = height

  def area(self):
    ar = (self.base * self.height) / 2
    return ar
```

```python
circle1 = Circle(5)
square1 = Square(4)
triangle1 = Triangle(6, 3)
```

```python
print("Circle area:", circle1.area(),"cm2")
print("Square area:", square1.area(),"cm2")
print("Triangle area:", triangle1.area(),"cm2")
```

```python
shapes = [circle1,square1,triangle1]
for s in shapes:
  print(f"The area of the shape is:", s.area(),"cm2")
```
